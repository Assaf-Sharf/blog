#### SEARCH 

```python
# 해당값이 배열에 있는지 모르기때문에 조든 원소를 다 뒤져야한다.
# random access 가 있어 메모리에 쉽게 접근할 수  있지만 그 안의 값은 모른다. 
# 배열의 값을 하나하나 체크하는 과정을 거쳐야한다.
# 그냥 읽는 것보다 시간이 걸린다 
# 최악의 경우에는 모든 데이터를 뒤졌지만 값이 없는 경우이다. 
# 위 상황을 선형검색이라고 한다. 


```

```python
#binary search
# 1 ~ 10으로 이루어진 배열 중 9 찾기 

cnt = 0

def binary_search(arr, num):
    global cnt 
    cnt +=1
    
    print(f"[{cnt}] : {arr}")
    if len(arr) == 1:
         return arr[0]
    
    mid = len(arr)//2
    print(f"mid : {mid}")
    if num == arr[mid]:
         return arr[mid]
    if num>arr[mid]:
         return binary_search(arr[mid:], num)
    else:
         return binary_search(arr[:mid], num)
        
```

```python
global cnt 
cnt =0

arr = [i for i in range(1, 11)]
num = 7
print(f"search num : {num}")
print(f"array : {arr}")
print('-'*(len(arr)*4))
binary_search(arr, num)
```

bubble sort

시간 복잡도 O(n^2)

```python
def bubble_sort(arr):
    global cnt
    cnt = 0
    for j in range(len(arr)):
        cnt+=1
        swap_cnt = 0
        for i in range(len(arr)-j-1):
            if arr[i] > arr[i+1]:
                swap_cnt+=1
                print(f'swap [{swap_cnt}] : {arr[i]} <-> {arr[i+1]}')
                arr[i], arr[i+1] = arr[i+1], arr[i]

        print(f"STEP [{cnt}] : {arr}")
        print('='*40)

    return arr
```

```python
arr = [6, 5, 4, 11, 7, 2, 1, 3]
print(f"arr : {arr}")
print('='*40)
bubble_sort(arr)
```

selection sort 선택 정렬

한번의 swap

그렇다면 선택정렬의 시간 복잡도는 버블 정렬보다 좋겠지?!

```python
def selection_sort(arr):
    global cnt
    cnt = 0
    for i in range(len(arr)):
        cnt+=1
        idx = i
        for j in range(i,len(arr)):
            if arr[i] > arr[j]:
                idx = j
        arr[i], arr[idx] = arr[idx], arr[i]
        print(f"swap[{cnt}] : {i} <-> {idx}")
        print(f"STEP[{cnt}] : {arr}")
        print('='*40)
        
    return arr

```

```python
print(f"arr : {arr}")
print('='*40)
arr = [6, 5, 4, 11, 7, 2, 1, 3]
selection_sort(arr)
```

```python
def insert_sort(arr):
    key_idx = int(input("key idx : "))
    key = arr[key_idx]
    global cnt
    cnt = 0
    for i in range(len(arr)-key_idx):
        cnt+=1
        swap_cnt = 0
        key = arr[key_idx]
        print(f"key : {key}")
        for j in range(key_idx,0,-1):
            if arr[j] < arr[j-1]:
                swap_cnt+=1
                print(f"swap[{swap_cnt}] : {arr[j]} <-> {arr[j-1]}")
                arr[j], arr[j-1] = arr[j-1], arr[j]
            else:
                continue
        key_idx +=1
        
        print(f"STEP [{cnt}] : {arr}")
        print('='*40)
        
    return arr
```

```python
arr = [6, 5, 4, 11, 7, 2, 1, 3]
print(f"arr : {arr}")
print('='*40)
insert_sort(arr)
```
